```python
#I want to display the date without the time
df_2023['INCIDENT_DATE'] = df_2023['INCIDENT_DATETIME'].dt.date
df_2023
```

```python
#let's make this less unweidly and reduce scrolling
df_2023 = df_2023[['INCIDENT_DATE','BOROUGH','INITIAL_CALL_TYPE','ZIPCODE','response_time_min']]
df_2023
```

```python
#let's see whether response times or the number of calls made are any different depending on the season
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# let's make a new column for the season the incident occurred in 
df_2023['Season'] = df_2023['INCIDENT_DATE'].apply(get_season)
df_2023
```

```python
#how many calls occurred in each season?
df_2023['Season'].value_counts()
```

```python
#how many calls occurred in each season and each borough?
seasonal_borough_counts = df_2023.groupby(['Season', 'BOROUGH']).size().reset_index(name='call_count')

print(seasonal_borough_counts)
```

```python
# What was the median response time in each borough in each season? (median)
median_response_time = df_2023.groupby(['Season', 'BOROUGH'])['response_time_min'].median().reset_index(name='MEDIAN_RESPONSE_TIME')
print(median_response_time)
```

```python
# What was the mean response time in each borough in each season? (mean)
median_response_time = df_2023.groupby(['Season', 'BOROUGH'])['response_time_min'].mean().reset_index(name='MEDIAN_RESPONSE_TIME')
print(median_response_time)
```
